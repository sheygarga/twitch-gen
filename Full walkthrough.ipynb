```python
import pandas as pd
import numpy as np
#import Levenshtein as lvs
import re
from textgenrnn import textgenrnn

textgen = textgenrnn()
```

```python
data = pd.read_json('CleanedData.txt', orient="columns")
data.head()
```

```python
test_list = []
body_list = []
for index, row in data.iterrows():
    body_list.append(row['message']['body'])
    test_list.append(row['message']['fragments'][0]['text'])
```

```python
np.savetxt('test-fragment.txt', test_list, fmt='%s', newline = ',', encoding='utf-8')
np.savetxt('test-body.txt', body_list, fmt='%s', newline = ',', encoding='utf-8')
```

```python
textgen.reset()
textgen.train_on_texts(body_list, gen_epochs=2)
```

```python
textgen.generate(5)
```
